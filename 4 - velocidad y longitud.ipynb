# Velocidad de fase y longitud de atenuación

## Bibliotecas

```python
import matplotlib.pyplot as plt
import scienceplots
import seaborn as sns
import pandas as pd
import numpy as np
```

## Opciones de graficación
Estética de las gráficas

```python
sns.set_context("paper", font_scale = 1.6)
plt.style.use(["science", "grid"])
```

## Funciones

```python
def rekd(a, b, freq, norma):
    arcotan = np.arctan(-2 * b * freq / (a ** 2))
    cc = norma * np.cos(arcotan)
    
    return cc / (2 * b)
```

```python
def imkd(a, b, freq, norma):
    arcotan = np.arctan(-2 * b * freq / (a ** 2))
    dd = norma * np.sin(arcotan)
    
    return (dd + a) / (2 * b)
```

```python
def reki(a, b, freq, norma):
    arcotan = np.arctan((2 * b * freq / (a ** 2)) + np.pi)
    cc = norma * np.cos(arcotan)

    return cc / (2 * b)
```

```python
def imki(a, b, freq, norma):
    arcotan = np.arctan((2 * b * freq / (a ** 2)) + np.pi)
    dd = norma * np.sin(arcotan)

    return (dd - a) / (2 * b)
```

```python
frecuencias = np.linspace(0, 1, 10000)
```

```python
a = -0.480
b = 0.113
norma = ((a ** 4) + (16 * (b ** 2) * (frecuencias ** 2))) ** 0.25
```

```python
rkd = rekd(a, b, frecuencias, norma)
ikd = imkd(a, b, frecuencias, norma)
rki = reki(a, b, frecuencias, norma)
iki = imki(a, b, frecuencias, norma)
```

```python
plt.figure(figsize = (8, 4))
plt.xlim(0, 1)
plt.xlabel("$\omega$")
plt.ylabel("$V_p$")
plt.plot(frecuencias, frecuencias / rkd, label = "Direct")
plt.plot(frecuencias, frecuencias / rki, label = "Reverse")
plt.legend(loc = "best")
plt.savefig("graficas/velocidad.pdf", bbox_inches = "tight")
plt.show()
```

```python
plt.figure(figsize = (8, 4))
plt.xlim(0, 1)
plt.xlabel("$\omega$")
plt.ylabel("$l_a$")
plt.plot(frecuencias, np.abs(1 / ikd), label = "Direct")
plt.plot(frecuencias, np.abs(1 / iki), label = "Reverse")
plt.legend(loc = "best")
plt.savefig("graficas/longitud_corre.pdf", bbox_inches = "tight")
plt.show()
```
