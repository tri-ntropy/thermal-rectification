## Archivos con los datos del flujo de calor

```python
frecuencias = np.loadtxt("datos/frecuencias_linea.csv")
frecuencias.size
```

```python
exponentes = np.loadtxt("datos/exponente_linea.csv").astype(int)
exponentes.size
```

```python
flujocalor_izq = np.loadtxt("datos/flujocalor_izq_linea.csv")
flujocalor_izq.shape
```

```python
flujocalor_der = np.loadtxt("datos/flujocalor_der_linea.csv")
flujocalor_der.shape
```

```python
if np.all(flujocalor_der > flujocalor_izq):
    flujocalor_max = flujocalor_der
    flujocalor_min = flujocalor_izq
else:
    flujocalor_max = flujocalor_izq
    flujocalor_min = flujocalor_der
```

```python
flujocalor = flujocalor_max - flujocalor_min
flujocalor
```

```python
plt.figure(figsize = (8, 4))
plt.xlabel("$\omega$")
plt.ylabel("$R$")
plt.xscale("log")
for e in range(exponentes.size):
    plt.plot(frecuencias, flujocalor[e], label = f"$n = {exponentes[e]}$")
plt.legend(loc = "best")
plt.savefig("graficas/rectificacion_linea.pdf", dpi = 300, bbox_inches = "tight")
plt.show()
```
